```python
download_area_dir="../regions/shapes/bharatpur.geojson"
area_gdf=gpd.read_file(download_area_dir)   
#plot the area
m = leafmap.Map()
m.add_gdf(area_gdf, layer_name="Area of Interest",zoom_to_layer=True,style={'fillColor': '#00000000', 'color': 'red'})
# m
```

```python
import folium
from shapely.wkt import loads
from shapely.geometry import mapping

# Define the WKT polygon for Delhi
delhi_wkt = 'POLYGON((77.0689 28.4038, 77.0689 28.8836, 77.3418 28.8836, 77.3418 28.4038, 77.0689 28.4038))'

# Convert WKT to a shapely geometry
delhi_polygon = loads(delhi_wkt)

# Create a folium map centered around Delhi
m = folium.Map(location=[28.6437, 77.2056], zoom_start=11)

# Add the polygon to the map
folium.GeoJson(
    data=mapping(delhi_polygon),
    style_function=lambda x: {'fillColor': 'transparent', 'color': 'red'}
).add_to(m)

# Display the map
# m
```

```python
# area_of_interest = geojson_to_wkt(read_geojson(download_area_dir))
start_date = '20241201'
end_date = '20241231'
platformname = 'Sentinel-5P'
producttype = 'L2__CO___'
products = api.query(delhi_wkt,
                     date=(start_date, end_date),
                     platformname=platformname,
                     producttype=producttype)

# Display number of products found
print(f"Number of products found: {len(products)}")
```

### In this notebook I found top n brick kilns tiles which have the highest number of brick kilns in them. 

```python
import os
```

```python
def get_files_with_most_labels(directory):
    labels_count = {}
    for filename in os.listdir(directory):
        # print(filename)
        if filename.endswith(".txt"):
            file_path=os.path.join(directory, filename)
            with open(file_path, "r") as file:
                label_count=len(file.readlines())
                labels_count[filename] = label_count

    sorted_files = sorted(labels_count.items(), key=lambda x: x[1], reverse=True)
    if sorted_files:
        most_labels_file = sorted_files[0][0]
        return most_labels_file,sorted_files
    else:
        return None, []
                
```

```python
state_name = 'punjab'
directory = f'/home/patel_zeel/kiln_compass_24/data/{state_name}/labels'

most_labels_file, sorted_files = get_files_with_most_labels(directory)

if most_labels_file:
    print(f"File with the most labels: {most_labels_file}")
    print("All files sorted by label count in descending order:")
    for file, count in sorted_files:
        print(f"{file}: {count} labels")
else:
    print("No label files found.")
```

```python
# selct top 15 files with most labels and copy them to another directory
top_files = sorted_files[:15]
copy_dir=f"../data/vlm_data/{state_name}_most_15/labels/"
os.makedirs(copy_dir, exist_ok=True)
for file, count in top_files:
    src_file_path=os.path.join(directory, file)
    dst_file_path=os.path.join(copy_dir, file)
    os.system(f"cp {src_file_path} {dst_file_path}")
```

```python
# state_name="lucknow_airshed"
destination_region=f'{state_name}_most_15'
label_dir = f"../data/vlm_data/{state_name}_most_15/labels"
source_image_dir = f"/home/patel_zeel/kiln_compass_24/data/{state_name}/images"
destination_image_dir = f"../data/vlm_data/{destination_region}/images"

# Create destination directory if it doesn't exist
os.makedirs(destination_image_dir, exist_ok=True)

# Iterate over all .txt files in the label directory
for label_file in os.listdir(label_dir):
    if label_file.endswith(".txt"):
        # Extract the base filename (without extension)
        base_name = os.path.splitext(label_file)[0]
        # Construct the corresponding .tif filename
        tif_filename = f"{base_name}.tif"
        # Construct the source .tif path
        source_tif_path = os.path.join(source_image_dir, tif_filename)
        # Construct the destination symlink path
        destination_tif_path = os.path.join(destination_image_dir, tif_filename)

        # Check if the corresponding .tif exists in the source directory
        if os.path.exists(source_tif_path):
            # Create a symbolic link in the destination directory
            if not os.path.exists(destination_tif_path):  # Avoid overwriting existing symlinks
                os.symlink(source_tif_path, destination_tif_path)
                print(f"Linked {source_tif_path} -> {destination_tif_path}")
            else:
                print(f"Symlink already exists for {destination_tif_path}")
        else:
            print(f"File not found: {source_tif_path}")

print(f"All matching .tif files have been symlinked to {destination_image_dir}")
```

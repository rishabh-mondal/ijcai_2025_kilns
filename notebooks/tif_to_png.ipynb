```python
import os
from PIL import Image

# Directory containing the .tif files
input_dir = '/home/rishabh.mondal/Brick-Kilns-project/ijcai_2025_kilns/data/gms_data/delhi_ncr_data/images'

# Directory to save the .png files
output_dir = '/home/rishabh.mondal/Brick-Kilns-project/ijcai_2025_kilns/data/gms_data/delhi_ncr_data/converted_images'

# Create the output directory if it doesn't exist
os.makedirs(output_dir, exist_ok=True)

# Loop through each file in the input directory
for filename in os.listdir(input_dir):
    if filename.endswith('.tif'):
        # Construct full file path
        filepath = os.path.join(input_dir, filename)
        
        # Open the .tif image
        with Image.open(filepath) as img:
            # Convert the filename to .png
            png_filename = filename.replace('.tif', '.png')
            png_filepath = os.path.join(output_dir, png_filename)
            
            # Save the image in .png format to the output directory
            img.save(png_filepath, 'PNG')
        
        print(f"Converted {filename} to {png_filename}")

print("Conversion completed. Files are saved in:", output_dir)
```

```python
import os
import shutil

# Define paths
labels_dir = '/home/rishabh.mondal/Brick-Kilns-project/ijcai_2025_kilns/data/gms_data/delhi_ncr_data/train_obb_data/labels_1'
images_dir = '/home/rishabh.mondal/Brick-Kilns-project/ijcai_2025_kilns/data/gms_data/delhi_ncr_data/converted_images'
output_dir = '/home/rishabh.mondal/Brick-Kilns-project/ijcai_2025_kilns/data/gms_data/delhi_ncr_data/train_obb_data/images_1'

# Create output directory if it doesn't exist
os.makedirs(output_dir, exist_ok=True)

# Get the list of label filenames (without extension)
label_filenames = [os.path.splitext(f)[0] for f in os.listdir(labels_dir) if f.endswith('.txt')]

# Loop through each label filename and copy the corresponding image
for label_name in label_filenames:
    image_filename = f"{label_name}.png"  # Assuming images are in .tif format
    image_path = os.path.join(images_dir, image_filename)
    
    # Check if the image exists
    if os.path.exists(image_path):
        shutil.copy(image_path, output_dir)
        print(f"Copied {image_filename} to {output_dir}")
    else:
        print(f"Image {image_filename} not found in {images_dir}")

print("Image copying completed.")
```

```python
#read the geojson file
delhi_data=gpd.read_file('../regions/shapes/ncr_brick_kilns.geojson')
delhi_data.head(2)
```

```python
for index, row in delhi_data.iterrows():
    bbox = [row['min_lon'], row['min_lat'], row['max_lon'], row['max_lat']]
    # print(bbox)
    filename = f"../data/gms_data/delhi_ncr_brick_kilns/{row['center_lat']}_{row['center_lon']}.tif"

    leafmap.map_tiles_to_geotiff(filename,bbox, zoom=19,source='Google Satellite')


```

```python
import ee
import folium

# Initialize the Earth Engine library
ee.Initialize()

# Define the dataset
dataset = ee.ImageCollection('MODIS/061/MOD11A1') \
            .filter(ee.Filter.date('2024-06-01', '2025-01-10'))

# Select the 'LST_Day_1km' band for land surface temperature
land_surface_temperature = dataset.select('LST_Day_1km')

# Define visualization parameters
land_surface_temperature_vis = {
    'min': 13000.0,
    'max': 16500.0,
    'palette': [
        '040274', '040281', '0502a3', '0502b8', '0502ce', '0502e6',
        '0602ff', '235cb1', '307ef3', '269db1', '30c8e2', '32d3ef',
        '3be285', '3ff38f', '86e26f', '3ae237', 'b5e22e', 'd6e21f',
        'fff705', 'ffd611', 'ffb613', 'ff8b13', 'ff6e08', 'ff500d',
        'ff0000', 'de0101', 'c21301', 'a71001', '911003'
    ]
}

# Create a map centered on Lucknow, Uttar Pradesh
map_center = folium.Map(location=[26.8467, 80.9462], zoom_start=10)

# Get the map ID and token for the land surface temperature layer
map_id_dict = land_surface_temperature.getMapId(land_surface_temperature_vis)

# Add the Earth Engine layer to the map
folium.TileLayer(
    tiles=map_id_dict['tile_fetcher'].url_format,
    attr='Google Earth Engine',
    name='Land Surface Temperature',
    overlay=True,
    control=True
).add_to(map_center)

# Add a layer control to toggle the visibility of the layer
folium.LayerControl().add_to(map_center)

# Display the map
map_center
```

```python
import ee
import folium

# Initialize the Earth Engine library
ee.Initialize()

# Read in the NEON AOP RGB Camera Image Collection
rgb = ee.ImageCollection("projects/neon-prod-earthengine/assets/RGB/001")

# Display available images in the RGB/001 Image Collection
print('NEON RGB Camera Images:', rgb.aggregate_array('system:index').getInfo())

# Specify the start and end dates and filter by date range
start_date = ee.Date('2021-01-01')
end_date = start_date.advance(1, 'year')
rgb_2021 = rgb.filterDate(start_date, end_date)

# Filter by NEON site name (replace with actual site property if needed)
rgb_abby_2021 = rgb_2021.filter(ee.Filter.eq('NEON_SITE', 'ABBY')).mosaic()

# Define visualization parameters
vis_params = {
    'min': 40,
    'max': 200,
    'gamma': 0.65
}

# Create a map centered on the ABBY site
map_center = folium.Map(location=[45.75, -122.341], zoom_start=19)

# Get the map ID and token for the RGB Camera layer
map_id_dict = ee.Image(rgb_abby_2021).getMapId(vis_params)

# Add the RGB Camera layer to the map
folium.TileLayer(
    tiles=map_id_dict['tile_fetcher'].url_format,
    attr='Google Earth Engine',
    name='ABBY 2021 RGB Camera Imagery',
    overlay=True,
    control=True
).add_to(map_center)

# Add a layer control to toggle the visibility of the layer
folium.LayerControl().add_to(map_center)

# Display the map
map_center
```

```python
import ee
import folium

# Initialize the Earth Engine library
ee.Initialize()

# Define the dataset for surface air temperature
dataset = ee.ImageCollection('NCEP_RE/surface_temp') \
            .filter(ee.Filter.date('2018-08-01', '2018-08-15'))

# Select the 'air' band for surface air temperature
surface_air_temperature = dataset.select('air')

# Define visualization parameters for the surface air temperature
surface_air_temperature_vis = {
    'min': 190.0,
    'max': 308.0,
    'palette': [
        '800080', '0000ab', '0000ff', '008000', '19ff2b', 'a8f7ff', 'ffff00',
        'd6d600', 'ffa500', 'ff6b01', 'ff0000'
    ],
}

# Define a map centered on Uttar Pradesh (latitude and longitude for UP)
map_center = folium.Map(location=[26.8467, 80.9462], zoom_start=7)

# Get the map ID and token for adding the layer
map_id = surface_air_temperature.getMapId(surface_air_temperature_vis)

# Add the surface air temperature layer to the map
folium.TileLayer(
    tiles=map_id['tile_fetcher'].url_format,
    attr='Google Earth Engine',
    name='Surface Air Temperature',
    overlay=True,
    control=True
).add_to(map_center)

# Add layer control to the map
map_center.add_child(folium.LayerControl())

# Display the map
map_center
```
